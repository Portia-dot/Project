```python
#Import the required dataset needed 
import os
import pandas as pd
import sys
from IPython.display import display
from dis import dis
import geopandas
import pandas as pd
from shapely.wkt import loads
```

```python
#Get the data from our folder and clean them up FOR area of burnt

path_AoB = []
for i in range(0, 10):
    path = os.path.abspath(os.path.join(os.getcwd(),
                                        '../Wildfires_Date_of_Burning_1559/data/ABoVE_DoB_201' + str(i)  + '.shp' ))
    path_AoB.append(path)
```

```python
#Fire Charateristics Dataset
charateristics = []
for i in range(0, 10):
    path = os.path.abspath(os.path.join(os.getcwd(), '../Wildfires_Date_of_Burning_1559/data/ABoVE_DoB_MODIS_Active_Fires_201' + str(i)  + '.shp' ))
    charateristics.append(path)
```

```python
def make_dataframes(aob_path, chracteristics_path):
    gdf_extents = []
    for path in aob_path:
        gdf_extents.append(geopandas.GeoDataFrame.from_file(path))
    
    gdf_locations = []
    for path in chracteristics_path:
        gdf_locations.append(geopandas.GeoDataFrame.from_file(path))

    gdf_aob = pd.concat(gdf_extents, axis = 0)
    gdf_characteristics = pd.concat(gdf_locations, axis = 0)

    # filter merged data for Alberta and BC
    gdf_aob_AB_BC = gdf_aob[(gdf_aob["REF_ID"].apply(lambda s : s[0:2] == "AB")) | 
                            (gdf_aob["REF_ID"].apply(lambda s : s[0:2] == "BC"))]

    gdf_characteristics_AB_BC = gdf_characteristics[(gdf_characteristics["REF_ID"].apply(lambda s : s[0:2] == "AB")) | 
                            (gdf_characteristics["REF_ID"].apply(lambda s : s[0:2] == "BC"))]

    return gdf_aob_AB_BC, gdf_characteristics_AB_BC
```

```python
#Create the datadrame from our shapefiles
AoB,Charateristics = make_dataframes(path_AoB,  charateristics)
```

```python
path_AoB = os.path.abspath(os.path.join(os.getcwd(), '../Wildfires_Date_of_Burning_1559/data/AoB.csv'))
charateristics =  os.path.abspath(os.path.join(os.getcwd(), '../Wildfires_Date_of_Burning_1559/data/Charateristics.csv'))

AoB.to_csv(path_AoB, index = False)
Charateristics.to_csv(charateristics, index = False)
```
