```python
def access_dataframe(dataframe):
    #convert date to datetime
    dataframe['date'] = pd.to_datetime(dataframe['date'])
    #extract day, month, year, time
    dataframe['day'] = dataframe['date'].dt.day
    dataframe['month'] = dataframe['date'].dt.month
    dataframe['year'] = dataframe['date'].dt.year
    dataframe['time'] = dataframe['date'].dt.time
    #extract season
    dataframe['season'] = np.where(dataframe['month'].isin([3,4,5]), 'spring',
                    np.where(dataframe['month'].isin([6,7,8]), 'summer',
                    np.where(dataframe['month'].isin([9,10,11]), 'autumn', 'winter')))
    #check if day is on the weekend or weekday
    dataframe['weekday'] = np.where(dataframe['date'].dt.dayofweek.isin([5,6]), 'weekend', 'weekday')

    return dataframe
```

```python
to_drop = ['lon', 'lat', 'lon_std','Status', 'FRCNTR']
WeatherData.rename(columns = {'distanceinkm':'distance_to_firecenter'}, inplace = True)
```

```python
categorical_variables = ['FIRE_CAUSE', 'sat','season', 'weekday']
label = 'AoB_Category'
continuous  = [col for col in WeatherData.columns if col not in to_drop + [label] + categorical_variables]
```

```python
print(continuous)
```

```python
# set indicators for categorical data
categorical_features = pd.get_dummies(WeatherData[categorical_variables].astype('str'))
categorical_features.head()
```

```python
#Normalize all the continuos variables
X_all = WeatherData[continuous].to_numpy()
continuous_all_df = pd.DataFrame(X_all, columns=continuous)
features_all = pd.concat([continuous_all_df, categorical_features], axis = 1)
display(features_all.head())
display(features_all.shape)
```

```python
no_stds = [column for column in continuous if column.find('std') == -1]
X_no_stds = WeatherData[no_stds]
continuous_no_stds_df = pd.DataFrame(X_no_stds, columns=no_stds)
features_no_stds = pd.concat([continuous_no_stds_df, categorical_features], axis=1)
display(features_no_stds.head())
display(features_no_stds.shape)
```

```python
# Encoding label data
from sklearn.preprocessing import LabelEncoder
y = WeatherData[label]
le = LabelEncoder()
y_encoded = le.fit_transform(y)
label_df = pd.DataFrame(y_encoded, columns=[label])
display(list(map(lambda x, y: x + ' --> ' + str(y), le.classes_, le.transform(le.classes_))))
```

```python
XY_all = pd.concat([features_all, label_df], axis=1)
XY_no_stds = pd.concat([features_no_stds, label_df], axis=1)
display(XY_all.shape, XY_no_stds.shape)
```

```python
#Save the dataset with stds
XY_all.to_csv('PredictorFinalWithSTDS.csv',index = False )
XY_no_stds.to_csv('PredictorFinalWithOutSTDS.csv',index = False )
```
