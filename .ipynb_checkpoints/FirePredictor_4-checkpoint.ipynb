```python
#Import all the required dependencies
import os 
import sys
import pandas as pd
sys.path.insert(1, os.path.join(os.getcwd(), "..", ""))
import xarray as xr
from netCDF4 import Dataset
import numpy as np
```

```python
path =  os.path.abspath(os.path.join(os.getcwd(), '../NewDataset.nc'))
```

```python
import xarray as xr
DS = xr.open_dataset(path)
DS.to_dataframe().to_csv("Dataset.csv")
```

```python
#Function
def get_rounded_locations(df):
    df_lat_rounded = df.lat.map(lambda x: round(x * 4) / 4)
    df_lon_rounded = df.lon.map(lambda x: round(x * 4) / 4)
    return df_lat_rounded, df_lon_rounded
```

```python
path =  os.path.abspath(os.path.join(os.getcwd(), '../Dataset/Dataset.csv'))
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
path_fires = os.path.abspath(os.path.join(os.getcwd(), '../charateristics.csv'))
```

```python
rounddata = pd.read_csv(path_fires, low_memory = False)
```

```python
df_lat_rounded, df_lon_rounded = get_rounded_locations(rounddata)
```

```python
rounddata.head()
```

```python
#Convert to datatime
data['Date'] = pd.to_datetime(data['time'])
```

```python
data.head()
```

```python
data = data[(data["latitude"].isin(set(df_lat_rounded))) & (data["longitude"].isin(set(df_lon_rounded)))]
```

```python
data.head()
```

```python
data.info()
```

```python
#Generate WindSpeed 
from numpy import sqrt
data['wind_speed'] = data.v10**2 + data.u10**2
```

```python
data.head()
```

```python
data['wind_speed'] = data['wind_speed'].apply(lambda x: sqrt(x))
```

```python
data.head()
```

```python
#Drop u10 and v10
data.drop(['u10', 'v10'], axis = 1, inplace = True)
```

```python
##Save to a dataset
data.to_csv('ProcessedDataset.csv')
```
